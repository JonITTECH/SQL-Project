```python
##solutions_file.csv
import pandas as pd
data = pd.read_csv('solutions_file.csv')
data
```

```python
data_sorted = data.sort_values(by='result', ascending=False)
data_sorted.head(10)
```

```python
data_sorted[data_sorted['school_name']=='ironhack']
```

```python
data_sorted[data_sorted['country']!='United States'][['school_name','result','flag_online_school','flag_flexible_school','flag_job_school','flag_outcomes_school','flag_bill_school','flag_vet_school']].head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import numpy as np

plt.figure(figsize=(6, 4))
plt.boxplot(data_sorted['result'], vert=False)  # Create a horizontal box plot
q1 = data_sorted['result'].quantile(0.25)
q2 = data_sorted['result'].quantile(0.5)
q3 = data_sorted['result'].quantile(0.75)

print('First quartile:',q1)
print('Second quartile:',q2)
print('Third quartile:',q3)
plt.title('Box Plot of result metric')
plt.xlabel('Result')
plt.show()

plt.figure(figsize=(8, 6))
sns.distplot(data_sorted['result'], kde=True, label='Male', color='skyblue')
```
